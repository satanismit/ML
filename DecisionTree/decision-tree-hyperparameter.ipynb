```python
import seaborn as sns 
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
```

```python
df= pd.read_csv('/kaggle/input/titanic-dataset/Titanic-Dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
X=df.drop(columns=['Survived','PassengerId','Name','Cabin','Ticket'])
y=df['Survived']
```

```python
X.info()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
X['Sex'] = le.fit_transform(X['Sex'])           # male → 1, female → 0
X['Embarked'] = le.fit_transform(X['Embarked']) # S, C, Q → 2, 0, 1
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.25, random_state=42)
```

Find the Best Parameter for Decision Tree 

```python
params = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None]
}

dt = DecisionTreeClassifier(random_state=42)
grid = GridSearchCV(dt, params, cv=10, scoring='accuracy')
grid.fit(X_train, y_train)

print("Best Params:", grid.best_params_)
print("Best Score:", grid.best_score_)
```
